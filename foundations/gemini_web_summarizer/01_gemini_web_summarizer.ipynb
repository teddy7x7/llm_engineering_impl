1. System Prompt  
    This is a predefined "behavior guideline" or "identity definition" for the model before the conversation begins. It is not part of the conversation content but is the underlying framework for model operation.
    * Role definition: Defines the model's personality, professional field, and tone.
    * Constraints: Dictates what the model is not allowed to do (e.g., avoid mentioning competitors, only respond in traditional Chinese).
    * Output format: Requires the model's structure for output (e.g., always return in JSON format, must use Markdown tables).
    * Persistence: The model remembers this setting throughout the entire conversation (Chat Session).
        
        Example: "You are a senior Python engineer, please remain professional and calm when replying, code examples must include detailed comments, and try to keep replies concise for simple questions."
        

```python
# Define system prompt

system_prompt_humor = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

system_prompt = """
You are a professional assistant that analyzes the contents of a website,
and provides a comprehensive summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

2. User Prompt  
    This is the specific instruction or question entered by the end-user in the dialog box.
    * Role definition: Provides specific tasks or current data.
    * Variability: Changes at any time, with different goals for each round of conversation.
    * Trigger mechanism: It is the direct driver for the model to produce output.  
        Example: "Please help me write a script in Python to fetch webpage titles."

```python
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website content.
If it includes news or announcements, then summarize these too.
"""
```

```python
def prompt_for(website: str) -> str:
    return f"""
{user_prompt_prefix}
{website}
""".strip()
```

```python
def summarize(url: str) -> str:
    website = fetch_website_contents(url)

    model = genai.GenerativeModel(model_name="gemini-2.5-flash", system_instruction=system_prompt)

    response = model.generate_content(
        prompt_for(website)
    )

    return response.text

def display_summary(url: str):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://ihower.tw/blog/")
```
